```python
path = "/content/Cars.mp4"
```

```python
import ultralytics
from ultralytics import YOLO
import cv2
```

```python
model = YOLO('yolov8n.pt')
```

```python
cap = cv2.VideoCapture(path)
assert cap.isOpened(), 'Error reading video file'
```

```python
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
```

```python
#create a videowriter object to save the output to save output video with object Tracking
video_write = cv2.VideoWriter('object_tracking.avi',cv2.VideoWriter_fourcc(*'mp4v'),fps,(w,h))
```

```python
import ultralytics
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

# loop through each frame
while cap.isOpened():
    success, frame = cap.read()
    #check if fram is successfully read
    if not success:
        print('Video fram is empty or video processing has been successfully completed')
        break
    #object Tracking
    track = model.track(frame, persist=True, tracker='bytetrack.yaml') #default: Bot-sort
    annotated_fram = track[0].plot()
    video_write.write(annotated_fram)
    # cv2_imshow(annotated_fram) # Changed cv2.imshow to cv2_imshow
cap.release()
video_write.release()
```
